```python
# # function for text preprocessing 
# def transform_text(text):
#     text = text.lower()
#     text = nltk.word_tokenize(text)
    
#     y = []
#     for i in text:
#         if i.isalnum():
#             y.append(i)
    
#     text = y[:]
#     y.clear()
    
#     for i in text:
#         if i not in stopwords.words('english') and i not in string.punctuation:
#             y.append(i)
            
#     text = y[:]
#     y.clear()
    
#     for i in text:
#         y.append(ps.stem(i))
    
            
#     return " ".join(y)
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import string

# Initialize Porter Stemmer
ps = PorterStemmer()

def transform_text(text):
    if isinstance(text, str):  # Check if text is a string
        # Convert text to lowercase
        text = text.lower()

        # Tokenize text
        text = nltk.word_tokenize(text)

        # Remove non-alphanumeric characters
        text = [word for word in text if word.isalnum()]

        # Remove stopwords and punctuation
        text = [word for word in text if word not in stopwords.words('english') and word not in string.punctuation]

        # Apply stemming
        text = [ps.stem(word) for word in text]

        # Join the words
        return " ".join(text)
    else:
        return ''  # Return an empty string for missing values
```

```python
# applying the text transformation to our training data
df['reviews.text'].apply(transform_text)
```

```python
df['transform_text'] = df['reviews.text'].apply(transform_text)
```

## Model Building

```python
#text Vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
X_train = cv.fit_transform(df['transform_text']).toarray()
```

```python
X_train.shape
```

```python
y_train=df['target'].values
```

```python
y_train.shape
```

## Trying different Naive bayes nodels

```python
#importing NB models
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score
```

```python
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

```python
# Gaussian NB model accuracy, precision, f1, roc_auc_score

gnb.fit(X_train,y_train)
y_pred1 = gnb.predict(X_train)
print('Accuracy ',accuracy_score(y_train,y_pred1))
print(confusion_matrix(y_train,y_pred1))
print(precision_score(y_train,y_pred1))
print(recall_score(y_train,y_pred1))
print(f1_score(y_train,y_pred1))
print(roc_auc_score(y_train,y_pred1))
```

```python
# Multinomial NB model accuracy, precision, f1, roc_auc_score

mnb.fit(X_train,y_train)
y_pred2 = mnb.predict(X_train)
print(accuracy_score(y_train,y_pred2))
print(confusion_matrix(y_train,y_pred2))
print(precision_score(y_train,y_pred2))
print(recall_score(y_train,y_pred2))
print(f1_score(y_train,y_pred2))
print(roc_auc_score(y_train,y_pred2))
```

```python
# Bernoulli NB model accuracy, precision, f1, roc_auc_score

bnb.fit(X_train,y_train)
y_pred3 = bnb.predict(X_train)
print(accuracy_score(y_train,y_pred3))
print(confusion_matrix(y_train,y_pred3))
print(precision_score(y_train,y_pred3))
print(recall_score(y_train,y_pred3))
print(f1_score(y_train,y_pred3))
print(roc_auc_score(y_train,y_pred3))

accuracy = accuracy_score(y_train, y_pred3)
precision = precision_score(y_train, y_pred3)
recall = recall_score(y_train, y_pred3)
f1 = f1_score(y_train, y_pred3)
roc_auc = roc_auc_score(y_train, y_pred3)

# Determine if the product has positive reviews based on evaluation metrics
if accuracy >= 0.7 and precision >= 0.7 and recall >= 0.7 and f1 >= 0.7 and roc_auc >= 0.6:
    print("The product has positive reviews in the market.")
else:
    print("The product does not have positive reviews in the market.")
```
